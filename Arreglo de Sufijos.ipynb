```python
palabra = "Recuperación$"
cadena = list(palabra.lower())
sufijos = {}
print(cadena)
for i in range(len(cadena)):
    temp ="".join(cadena[i:len(cadena)])
    sufijos[temp]= i+1
print("El texto contiene los siguientes sufijos:")
for sufijo in sufijos.items():
    print(sufijo)
print("Sufijos ordenados:")
ordenados = sorted(sufijos.items())
sufijosO = dict(ordenados)
for orden in ordenados:
    print(orden)
print(type(ordenados))
```
